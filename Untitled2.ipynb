```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import uuid
import os
import time


os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/colabModel/coldModel2/best.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    results = model(frame)
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
